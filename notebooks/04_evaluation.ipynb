# Model Evaluation
## Ethiopian Household Wealth Prediction

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
import sys, os, joblib, warnings
warnings.filterwarnings('ignore')
sys.path.append(os.path.abspath('..'))
from src.inference import WealthPredictorAPI
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
%matplotlib inline
```

```python
# Load data and model
X_test = pd.read_csv('../data/processed/X_test.csv')
y_test = pd.read_csv('../data/processed/y_test.csv').iloc[:, 0]
api = WealthPredictorAPI(model_path='../models/')
# Predict
y_pred_log = api.model.predict(X_test)
y_pred = np.expm1(y_pred_log)
y_true = np.expm1(y_test)
# Metrics
r2 = r2_score(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae = mean_absolute_error(y_true, y_pred)
print(f"R²: {r2:.4f}, RMSE: {rmse:,.0f}, MAE: {mae:,.0f}")
```

```python
# Actual vs Predicted
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(y_true, y_pred, alpha=0.3, s=5)
lims = [min(y_true.min(), y_pred.min()), max(y_true.max(), y_pred.max())]
axes[0].plot(lims, lims, 'r--'); axes[0].set_xlabel('Actual'); axes[0].set_ylabel('Predicted')
axes[0].set_title(f'Actual vs Predicted (R²={r2:.3f})')
residuals = y_true - y_pred
axes[1].hist(residuals, bins=50, edgecolor='black', alpha=0.7)
axes[1].axvline(0, color='r', linestyle='--'); axes[1].set_xlabel('Residual')
axes[1].set_title('Residual Distribution')
plt.tight_layout(); plt.show()
```
